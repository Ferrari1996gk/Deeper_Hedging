# Chiarella-Heston model calibrated to Nikkei 225 index data
# Use calibrated model to generate training dataset for deep hedging, and compare performance with delta hedging and GBM-based deep hedging

# 1. Data
### Nikkei 225  daily price data from 1999-07-08 to 2023-12-31, in total 6000 days. The first 3000 days are for calibrating the Chiarella-Heston model, GBM, and Extended Chiarella model.

```python
# Dependency
import yfinance as yf
import numpy as np
import pandas as pd
import pickle
import matplotlib.pyplot as plt
from statsmodels.graphics.tsaplots import plot_acf
np.random.seed(0)
# GBM model
def brownian_sim(num_path, num_period, mu, std, init_p, dt):
    z = np.random.normal(size=(num_path, num_period))

    a_price = np.zeros((num_path, num_period))
    a_price[:, 0] = init_p

    for t in range(num_period - 1):
        a_price[:, t + 1] = a_price[:, t] * np.exp(
            (mu - (std ** 2) / 2) * dt + std * np.sqrt(dt) * z[:, t]
        )
    return a_price
```

## Historical stylised facts - data from calibration

```python
YEAR = 250 # trading days
r = 0.03 #risk-free rate
```

```python
from helpers import hill_estimator
data = yf.download('^N225','1999-07-08','2011-09-29')
print('NUmber of days: %d' % len(data))
close = data['Adj Close'].values
his_ret = close[1:] / close[:-1] - 1
print("Hill estimator(measuring fat-tailed of returns): %.4f" % hill_estimator(his_ret.flatten()))
stdev = np.std(his_ret) * np.sqrt(YEAR)
print("Volatility: %.4f" % stdev)
drift = np.mean(his_ret) * YEAR
print("Average drift: %.4f" % drift)

fig, ax = plt.subplots(1, 1, figsize=(7, 3))
ax.plot(close)
ax.set_title('N225 close price for %d days' % len(data))
plt.show()


LAG = 30
fig, axs = plt.subplots(1, 2, figsize=(12, 4), sharey=True)
plot_acf(his_ret, ax=axs[0], lags=range(1, LAG), fft=True)
plot_acf(his_ret**2, ax=axs[1], lags=range(1, LAG), fft=True)
axs[0].set_ylim((-0.6, 0.6))
axs[0].set_xticks(range(1, LAG, 2))
axs[0].set_xlabel('Lag')
axs[0].set_title("Autocorrelation of Returns - N225 historical")
axs[1].set_xticks(range(1, LAG, 2))
axs[1].set_xlabel('Lag')
axs[1].set_title("Autocorrelation of Squared Returns - N225 historical")
plt.show()
```

## 3.1 GBM after calibration:

```python
a_price = brownian_sim(N, int(ttm / YEAR /dt), r, VOL, S, dt)
ret = a_price[:, 1:] / a_price[:, :-1] - 1
print("Hill estimator: %.4f" % hill_estimator(ret.flatten()))
stdev = np.mean(np.std(ret, axis=1) * np.sqrt(YEAR))
print("Average volatility: %.4f" % stdev)
drift = np.mean(ret) * YEAR
print("Average drift: %.4f" % drift)

fig, ax = plt.subplots(1, 1, figsize=(6, 4))
for i in range(100):
    ax.plot(a_price[i, :], alpha=0.3)
ax.set_title("Geometric Brownian Motion simulation")
plt.show()

x = ret[10]
from statsmodels.graphics.tsaplots import plot_acf
LAG = 30
fig, axs = plt.subplots(1, 2, figsize=(14, 4), sharey=True)
plot_acf(x, ax=axs[0], lags=range(1, LAG), fft=True)
plot_acf(x**2, ax=axs[1], lags=range(1, LAG), fft=True)
axs[0].set_ylim((-0.6, 0.8))
axs[0].set_xticks(range(LAG))
axs[0].set_title("autocorrelation of returns")
axs[1].set_xticks(range(LAG))
axs[1].set_title("autocorrelation of squared returns")
plt.show()
```

## 4.2 Deep hedging

### We use the Chiarella-Heston model to generate trading data for RL agent, and test the perfomance on the above testing dataset

### Two baseline: 1. traditional delta hedging; 2. GBM-generated training data

### All other parameters for the network and network trading related parameters (learning rate, batch_size, etc) are fixed and the same for both Chiarella-Heston and GBM

### The following are testing results after training for 10,000 episodes.

### The RL agent rewards details are attached in the "records" folder (in the .pkl files). 

```python
ROOT_DIR1 = "./records/abm0/" # performance data trained by GBM-based training set
ROOT_DIR2 = "./records/abm1/" # performance data trained by Chiarella-Heston based training set
mylist = []
tc_list = [0.001, 0.002, 0.004, 0.006, 0.008, 0.01] # transaction cost level, as percentage of transaction value
tag = 10 # We carry out testing after every 1000 episodes. Here 10 represents results after training for 10,000 episodes
for tc in tc_list:
    print("transaction cost: %.2f%%" % (tc*100))
    with open(ROOT_DIR1+'records_%.4f_%d.pkl' % (tc, tag), 'rb') as f:
        record = pickle.load(f)
        f.close()
        
    with open(ROOT_DIR2+'records_%.4f_%d.pkl' % (tc, tag), 'rb') as f:
        record2 = pickle.load(f)
        f.close()
    
    r_rl1 = np.nan_to_num(np.array(record['r_rl']))
    r_rl2 = np.nan_to_num(np.array(record2['r_rl']))
    r_delta1 = np.nan_to_num(np.array(record['r_delta']))
    r_delta2 = np.nan_to_num(np.array(record2['r_delta']))
    assert np.mean(r_delta1) == np.mean(r_delta2), "same traditional delta hedging must have same performance"
    print("Average profit and loss: ", np.round(np.mean(r_delta1), 2), np.round(np.mean(r_rl1), 2), np.round(np.mean(r_rl2), 2))
    print("Max loss: ", np.round(np.min(r_rl1), 4), np.round(np.min(r_rl2), 4))
    es1 = r_rl1[r_rl1 < np.quantile(r_rl1, 0.1)].mean()
    es2 = r_rl2[r_rl2 < np.quantile(r_rl2, 0.1)].mean()
    print("Expected shortfall: %.2f and %.2f" % (es1, es2))
    
    fig, ax = plt.subplots(figsize=(7, 4))
    # bins = np.arange(-1000, 500, 10)
    bins = 300
    res = ax.hist(r_rl1, bins=bins, alpha=0.5, label='DeepHedging - GBM', color='y')
    ymax = res[0].max()
    res = ax.hist(r_rl2, bins=bins, alpha=0.5, label='DeepHedging - Chiarella-Heston', color='r')
    ymax = max(ymax, res[0].max())
    res = ax.hist(r_delta1, bins=bins, alpha=0.5, label='Delta Hedging', color='black')
    ymax = max(ymax, res[0].max())
    ax.vlines(np.mean(r_rl1), ymin=0, ymax=ymax, color='y')
    ax.vlines(np.mean(r_rl2), ymin=0, ymax=ymax, color='r')
    ax.vlines(np.mean(r_delta1), ymin=0, ymax=ymax, color='black')

    ax.legend(loc='best')
    ax.set_title('Transaction Cost: %.4f%%'%(tc*100))
    ax.set_xlabel('P & L')
    ax.set_ylabel('Frequency')
    plt.show()
    mylist.append([np.mean(r_delta1), np.mean(r_rl1), np.mean(r_rl2)])
```

## The above figures show that with Transaction Cost increases, deep hedging performs better than tradtional delta hedging
